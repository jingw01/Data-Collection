
The purpose of this code is to load data from 800+ .htm files from various directories. 

In each htm file, there are *at least* 400 tables, and they are need to converted to .csv files. 

```python
#Import libaray and start the log

import requests
from pattern import web
import pandas as pd
import re
import os
import logging

#Start the log
log_filename='Scanhtm.log'
logging.basicConfig(filename=log_filename,
                    level=logging.DEBUG)
```

```python
"""
Function
---------
perdate

In each html file, there are multiples sub tables. 
This function is to extract value of a sub table:
 - variables that are the same across the table:
    'date, participant'
 - other variables in the table:
    'ID, Desc,Quantity,Price'

Parameter
---------
table：htm text of a subtab;e
"""

def perdate(table):
    m=re.search('FOR DATE: ',table)
    SettleDate=table[m.end():m.end()+10]
    m=re.search('\nPARTICIPANT: ',table)
    Participant=table[m.end():m.end()+5]

#CREATE A LIST OF POSITION OF LAST CHARACTER OF EACH VARIABLE 
#BY USING THE HEADER OF THE TABLE 
    tstart=table.index("\nID")+6
    tmiddle1=table.index("\n",tstart)
    tmiddle2=table.index("\n",tmiddle1+2)
    tend=table.index("\n",tmiddle2+2)
    trow=table[tmiddle2+2:tend]
    tp=re.compile("[A-Z] ")
    entry={}
    for i,m in enumerate(tp.finditer(trow)):
        entry[i]=m.end();

#EXTRACT THE DATA OF EACH ENTRY    
    p=re.compile('\nUS')
    start=[m.end() for m in p.finditer(table)]  #Identify the start of a row
    end=[m.end() for m in p.finditer(table)][1:] #Identify the end of a row
    end.append(len(table))
    data=[perrow(table,start[r],end[r],entry) for r in range(len(start)) ]
    data2=[dict(Date=SettleDate,
                Participant=Participant,
                ID=r[0],
                Desc=r[1],
                Quantity=r[2],
                Price=r[3]) for r in data]
    
    return pd.DataFrame(data2),len(data)
```

```python

"""
Function
---------
perrow

Each row of the data content the data for variables, ID, Desc, Price,
and Quantity. They are seperated by space. 

Since we know the position of last character of each variable, we would
be able to identify data for each variable.

Parameter
---------
table：htm text of a subtable;
start: the start position of the row;
end: the end position of the row;
entry: the list of position of last character for each variable
"""

def perrow(table,start,end,entry):
    row=table[start:end]
    ext=row.index('-',1)
    ID=row[0:ext].encode('utf8') # ID is identify if it is before a "-"
    middle=row.index("\n")       # ID and Desc are in a different row from pric and quantity;
    Desc=row[ext+2:middle].encode('utf8') 
    end=row.index("\n",middle+2)
    rowdata=row[middle+2:end] #limit to the second row
    Quantity=rowdata[0:entry[0]].encode('utf8')
    Price=rowdata[entry[0]+1:entry[1]].encode('utf8')
    result=[ID,Desc,Quantity,Price]
    
    return result

```

```python

"""
Function
---------
openfiles

Open an htm file and export all its table to one csv file

Parameter
---------
root: directory that the file is in
file: filename
"""


def openfiles(root,file):
    filename=os.path.join(root, file)
    logging.debug('Reading File:%s',file)
    pl=open(filename,"r")
    htm=pl.read()
    dom = web.Element(htm)
    
    #Create an empty dataframe

    cols=["Date","Participant","ID","Desc","Quantity"
          "Price"]
    df=pd.DataFrame(columns=cols)
    
    #Record Statistics of Each file, such the number of tables
    #in a page;
    totalrow=0
    count=0
    skipped=[]
    
    #Find all the tables and combine their values to one dataset

    for i,t in enumerate(dom.by_tag('pre')):
        table=t.content
        try:
            result,row=perdate(table)
            count=count+1
        except (ValueError,AttributeError) as e:
            result=pd.DataFrame()
            skipped.append(i)
            
        df=df.append(result,ignore_index=True)
    
    logging.debug("Skip Tag_Value: %s",skipped)
    logging.debug("Total number of %s Tables:",count)
    logging.debug("Total number of %s Entries:",totalrow)
    
    #Export the csv version of the file
    outname=file.replace(".htm",".csv")
    path=os.path.join('output\\',outname)
    df.to_csv(path)
    pl.close()
    return [outname,skipped,count,totalrow]
```

```python
#Read through several directories

recordt=["file","SkippedTag","NofTables","TotalRow"]
recorddf=pd.DataFrame(columns=recordt)
for root, dirs, files in os.walk("TopFolder"):

    logging.debug('Reading Folder: %s', root)
    for file in files:
        if file.endswith(".htm"):
            outname=file.replace(".htm",".csv")
            path=os.path.join('output\\',outname)
            if os.path.isfile(path):
                logging.debug('Already Imported File: %s', outname)
                print 'skip file: '+str(outname)
            else:
                print 'reading file: '+str(file)
                record=openfiles(root,file)
                tmpdf=pd.DataFrame(dict(zip(recordt,record)))
                recorddf=recorddf.append(tmpdf,ignore_index=True)

#Export Log Statistics
recorddf.to_csv("Record.csv")
```
